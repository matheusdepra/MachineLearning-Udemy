    Base de dados credito

```python
import pickle
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
from yellowbrick.classifier import ConfusionMatrix
```

```python
with open( './assets/credit.pkl', 'rb') as f:
    X_credit_treinamento, Y_credit_treinamento,    X_credit_teste, Y_credit_teste = pickle.load(f)
```

```python
#Com solver adam - padrao
rede_neural_credit = MLPClassifier(max_iter=1200, verbose=True, tol=0.00001 )
rede_neural_credit.fit(X_credit_treinamento, Y_credit_treinamento)
```

```python
#Com solver lbfgs - recomendado para datasets pequenos
rede_neural_credit_lbgfs = MLPClassifier(max_iter=2000, verbose=True, tol=0.00001, solver='lbfgs' )
rede_neural_credit_lbgfs.fit(X_credit_treinamento, Y_credit_treinamento)
```

```python
#Com solver adam - padrao
#Alterando para 2 camadas ocultas
rede_neural_credit2 = MLPClassifier(max_iter=2000, verbose=True, tol=0.00001,
                                  hidden_layer_sizes=(2,2))
rede_neural_credit2.fit(X_credit_treinamento, Y_credit_treinamento)
```

```python
#Rede neural default
previsoes = rede_neural_credit.predict(X_credit_teste)
accuracy_score(Y_credit_teste, previsoes)
```

```python
#Rede neural default - com lbgfs - treinamento +rapido
previsoes_lbgfs = rede_neural_credit_lbgfs.predict(X_credit_teste)
accuracy_score(Y_credit_teste, previsoes_lbgfs)
```

```python
#Rede neural 2 neuronios - +rapido e melhor resultado
previsoes2 = rede_neural_credit2.predict(X_credit_teste)
accuracy_score(Y_credit_teste, previsoes2)
```

```python
cm = ConfusionMatrix(rede_neural_credit2)
cm.fit(X_credit_treinamento, Y_credit_treinamento)
cm.score(X_credit_teste, Y_credit_teste)
```

```python
print(classification_report(Y_credit_teste, previsoes2))
```
